# Model Validation

## 1.Collinearity Check

#Calculate VIF = 1/(1-Rsquare) for all independent variables

```python
rsq_r=smf.ols('RDS~MKTS+ADMS',data=Startups_2).fit().rsquared
VIF_r=1/(1-rsq_r)
```

```python
rsq_m=smf.ols('MKTS~RDS+ADMS',data=Startups_2).fit().rsquared
VIF_m=1/(1-rsq_m)
```

```python
rsq_a=smf.ols('ADMS~RDS+MKTS',data=Startups_2).fit().rsquared
VIF_a=1/(1-rsq_a)
```

```python
# Putting these VIF values in Dataframe format
```

```python
D1={'Variables':['RDS','MKTS','ADMS'],'VIF':[VIF_r
,VIF_m,VIF_a]}
VIF_df=pd.DataFrame(D1)
VIF_df
```

## Model Deletion Diagnostics (checking Outliers)
 

### Cook's Distance Technique to check outliers

```python
# 1. Cook's Distance: If Cook's distance > 1, then it's an outlier
# Get influencers using cook's distance
(c,_)=Model.get_influence().cooks_distance
c
```

```python
# Plot the influencers using the stem plot
fig=plt.figure(figsize=(20,8))
plt.stem(np.arange(len(Startups_2)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# Index and value of influencer where C>0.5
np.argmax(c) , np.max(c)
```

```python
Startups_2[Startups_2.index.isin([49])] 
```

# Improve the  Model 

```python
#By deleting/droping the 49th observation from the data
```

```python
Startups_3=Startups_2.drop(Startups_2.index[[49]],axis=0).reset_index(drop=True)
Startups_3
```

# Model Deletion Diagnostics and Final Model

```python
while np.max(c)>0.5 :
    model=smf.ols("Profit~RDS+ADMS+MKTS",data=Startups_3).fit()
    (c,_)=model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    Startups_3=Startups_3.drop(Startups_3.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    Startups_3
else:
    final_model=smf.ols("Profit~RDS+ADMS+MKTS",data=Startups_3).fit()
    final_model.rsquared , final_model.aic
    print("Thus model accuracy is improved to",final_model.rsquared)
```
