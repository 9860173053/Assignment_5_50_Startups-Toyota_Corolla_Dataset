### 1.Collinearity Check

```python
## Calculate VIF = 1/(1-Rsquare) for all independent variables
```

```python
rsq_age=smf.ols('Age~KM+HP+CC+Doors+Gears+QT+Weight',data=Toyo3).fit().rsquared
Vif_age= 1/(1-rsq_age)
```

```python
rsq_KM=smf.ols('KM~Age+HP+CC+Doors+Gears+QT+Weight',data=Toyo3).fit().rsquared
Vif_KM= 1/(1-rsq_KM)
```

```python
rsq_HP=smf.ols('HP~KM+Age+CC+Doors+Gears+QT+Weight',data=Toyo3).fit().rsquared
Vif_HP= 1/(1-rsq_HP)
```

```python
rsq_CC=smf.ols('CC~KM+HP+Age+Doors+Gears+QT+Weight',data=Toyo3).fit().rsquared
Vif_CC= 1/(1-rsq_CC)
```

```python
rsq_Doors=smf.ols('Doors~KM+HP+CC+Age+Gears+QT+Weight',data=Toyo3).fit().rsquared
Vif_Doors= 1/(1-rsq_Doors)
```

```python
rsq_Gears=smf.ols('Gears~Age+KM+HP+CC+Doors+QT+Weight',data=Toyo3).fit().rsquared
Vif_Gears= 1/(1-rsq_Gears)
```

```python
rsq_QT=smf.ols('QT~Age+KM+HP+CC+Doors+Gears+Weight',data=Toyo3).fit().rsquared
Vif_QT= 1/(1-rsq_QT)
```

```python
rsq_Weight=smf.ols('Weight~KM+HP+CC+Doors+Gears+QT',data=Toyo3).fit().rsquared
Vif_Weight= 1/(1-rsq_Weight)
```
